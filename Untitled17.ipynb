```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
df = pd.read_csv('/content/drive/MyDrive/Mekong_Agri_CSV/CanTho_Full_Agri_2015_2024.csv')

df.head()
```

XỬ LÝ THỜI GIAN & TẠO FEATURE SEASONALITY

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date').reset_index(drop=True)

# Tạo feature mùa vụ
df['dayofyear'] = df['Date'].dt.dayofyear
df['sin_doy'] = np.sin(2*np.pi*df['dayofyear']/365)
df['cos_doy'] = np.cos(2*np.pi*df['dayofyear']/365)
```

CHỌN FEATURE & TARGET

```python
#ví dụ ta muốn dự báo nhiệt độ trung bình T2M
target_cols = ["GWETTOP", "RH2M", "PRECTOTCORR", "T2M"]

feature_cols = [
    "T2M", "T2M_MAX", "T2M_MIN",
    "RH2M", "PRECTOTCORR",
    "GWETTOP", "GWETROOT", "TS",
    "ALLSKY_SFC_SW_DWN", "EVPTRNS",
    "WS2M",
    "sin_doy", "cos_doy"
]

df_features = df[feature_cols].copy()
df_targets = df[target_cols].copy()
```

```python
#Chuẩn hóa dữ liệu
scaler_X = MinMaxScaler()
X_scaled = scaler_X.fit_transform(df_features)    # (N, num_features)

scaler_y = MinMaxScaler()
y_scaled = scaler_y.fit_transform(df_targets)
```

```python
#Tạo cửa sổ
LOOKBACK = 60
HORIZON = 1   # 1-step ahead

def create_lstm_dataset(features, targets_scaled, lookback):
    """
    features: (N, num_features)
    targets_scaled: (N, num_targets)  # ở đây num_targets = 4
    """
    X, y = [], []
    N = len(targets_scaled)
    for i in range(N - lookback):
        X_i = features[i : i + lookback, :]      # (60, num_features)
        y_i = targets_scaled[i + lookback, :]    # (4,)
        X.append(X_i)
        y.append(y_i)
    return np.array(X), np.array(y)

X, y = create_lstm_dataset(X_scaled, y_scaled, LOOKBACK)
print("X shape:", X.shape)  # (samples, 60, num_features)
print("y shape:", y.shape)  # (samples, 4)

# NGÀY ĐƯỢC DỰ ĐOÁN TƯƠNG ỨNG
# ============================
# Với sample i, ta dùng [i : i+60] để dự đoán ngày i+60
pred_dates = df["Date"].iloc[LOOKBACK:].reset_index(drop=True)
assert len(pred_dates) == len(y)
```

```python
# BUILD & TRAIN MODEL TRÊN TOÀN BỘ DATA
# ============================
num_features = X.shape[2]
num_targets = y.shape[1]  # = 4

model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(LOOKBACK, num_features)),
    Dropout(0.2),
    LSTM(32),
    Dropout(0.2),
    Dense(num_targets)    # output 4 giá trị: GWETTOP, RH2M, PRECTOTCORR, T2M
])

model.compile(optimizer="adam", loss="mse", metrics=["mae"])
model.summary()
```

```python
callbacks = [
    EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True),
    ModelCheckpoint("best_lstm_multitarget.h5", save_best_only=True, monitor="val_loss")
]

history = model.fit(
    X, y,
    validation_split=0.15,   # 15% sample cuối làm validation
    shuffle=False,           # rất quan trọng cho time series
    epochs=100,
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)
```

```python
# DỰ ĐOÁN CHO TOÀN BỘ (TRỪ 60 NGÀY ĐẦU)
# ============================
y_pred_scaled = model.predict(X)          # (n_samples, 4)
y_true_scaled = y                         # (n_samples, 4)

# inverse về giá trị thật (đơn vị gốc của từng cột)
y_pred_real = scaler_y.inverse_transform(y_pred_scaled)   # (n_samples, 4)
y_true_real = scaler_y.inverse_transform(y_true_scaled)   # (n_samples, 4)
```

```python
# LỌC 3 THÁNG GẦN NHẤT
# ============================
last_date = df["Date"].max()
start_last3 = last_date - pd.DateOffset(months=3)

mask_last3 = pred_dates >= start_last3

dates_last3 = pred_dates[mask_last3]
y_pred_last3 = y_pred_real[mask_last3]   # (n_last3, 4)
y_true_last3 = y_true_real[mask_last3]   # (n_last3, 4)

print("Ngày cuối cùng trong data:", last_date)
print("Bắt đầu 3 tháng gần nhất:", start_last3)
print("Số ngày trong 3 tháng gần nhất (có dự đoán):", len(dates_last3))
```

```python
# TÍNH MAE/RMSE TRONG 3 THÁNG GẦN NHẤT (CHO TỪNG BIẾN)
# ============================
print("\n[3 THÁNG GẦN NHẤT] MAE / RMSE cho từng biến:")
for i, col in enumerate(target_cols):
    mae_col = mean_absolute_error(y_true_last3[:, i], y_pred_last3[:, i])
    rmse_col = np.sqrt(mean_squared_error(y_true_last3[:, i], y_pred_last3[:, i]))
    print(f"  {col:12s} | MAE: {mae_col:.4f} | RMSE: {rmse_col:.4f}")

# Nếu muốn tính MAE/RMSE chung cho tất cả biến:
mae_all = mean_absolute_error(y_true_last3, y_pred_last3)
rmse_all = np.sqrt(mean_squared_error(y_true_last3, y_pred_last3))
print(f"\n[3 THÁNG GẦN NHẤT] MAE chung  (4 biến): {mae_all:.4f}")
print(f"[3 THÁNG GẦN NHẤT] RMSE chung (4 biến): {rmse_all:.4f}")
```

```python
# VẼ BIỂU ĐỒ 3 THÁNG GẦN NHẤT CHO 1–2 BIẾN QUAN TRỌNG
# ============================
# Ví dụ vẽ cho PRECTOTCORR (mưa) và T2M (nhiệt độ)
plot_vars = ["PRECTOTCORR", "T2M"]

for var in plot_vars:
    idx = target_cols.index(var)
    plt.figure(figsize=(14, 5))
    plt.plot(dates_last3, y_true_last3[:, idx], label=f"Thực tế {var}")
    plt.plot(dates_last3, y_pred_last3[:, idx], label=f"Dự báo {var}")
    plt.title(f"Dự báo {var} – 3 tháng gần nhất")
    plt.xlabel("Ngày")
    plt.ylabel(var)
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# IN VÍ DỤ 10 DÒNG CUỐI (3 THÁNG GẦN NHẤT)
# ============================
print("\nMột vài ngày trong 3 tháng gần nhất (thực vs dự báo):")
for d, true_row, pred_row in list(zip(dates_last3, y_true_last3, y_pred_last3))[:10]:
    print(f"{d.date()} | ", end="")
    for i, col in enumerate(target_cols):
        print(f"{col}: thực={true_row[i]:.3f} | dự={pred_row[i]:.3f}   ", end="")
    print()
```
